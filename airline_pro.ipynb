```python
# Q.4. Does price varies with airlines ?
data.head(2)
```

```python
 data_varies = data.groupby("airline")["price"].mean()
data_varies
```

```python
# Q.5. Does ticket price change based on the departure time and arrival time?
data.head(2)
```

```python
departure_time = data.groupby("departure_time")["price"].mean()
departure_time
```

```python
arrival_time = data.groupby("arrival_time")["price"].mean()
arrival_time
```

```python
sns.catplot(x="departure_time" , y="price" ,kind = "bar",data=data , palette ="rocket" , hue="class")
plt.show()

sns.catplot(x="arrival_time" , y="price" ,kind = "bar",data=data , palette ="rocket" , hue="class")
plt.show()
```

```python
sns.relplot(x="arrival_time" , y="price" ,kind = "line" , col="departure_time",data=data )
plt.show()
```
